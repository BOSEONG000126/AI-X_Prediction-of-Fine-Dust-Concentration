#### **<span style="color:blue">[1-5] 미세먼지 데이터 분석</span>**

* 데이터 전처리와 모델링을 위해 데이터 분석을 진행해보세요. 시각화 등을 통해 관측치들의 패턴 탐색, 잘못된 자료 탐색, 변수들간의 관계 파악을 할 수 있습니다.

```python
# 시각화 한글폰트 설정 (아래 코드를 실행하세요)

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')
```

```python
# air_21,air_22 의 다양한 변수에 대하여 시각화

from sklearn.preprocessing import LabelEncoder

# 라벨 인코더 초기화
label_encoder = LabelEncoder()

# 라벨 인코딩을 적용할 열 선택
categorical_columns = ['지역', '망', '측정소명', '주소']

# 라벨 인코딩을 적용할 데이터프레임 및 열 선택
for column in categorical_columns:
    air_21[column] = label_encoder.fit_transform(air_21[column])
    air_22[column] = label_encoder.fit_transform(air_22[column])

import matplotlib.pyplot as plt
import seaborn as sns

# 히트맵을 생성할 데이터프레임 (예제에서는 data_dropped를 사용)
colormap = plt.cm.PuBu
plt.figure(figsize=(40, 20))  # 그래프 크기 조절
plt.title("Correlation Heatmap", y=1.05, size=20)  # 제목 설정

# 상관 행렬을 히트맵으로 시각화
heatmap = sns.heatmap(air_22.astype(float).corr(), linewidths=0.1, vmax=1.0,
                      square=True, cmap=colormap, linecolor="white", annot=True, annot_kws={"size": 30})  # 레이블 크기 조절

# 가로 축 레이블 폰트 크기 조절
heatmap.set_xticklabels(heatmap.get_xticklabels(), size=15)

# 세로 축 레이블 폰트 크기 조절
heatmap.set_yticklabels(heatmap.get_yticklabels(), size=15)

plt.show()


```

#### **<span style="color:blue">[1-6] 날씨 데이터 분석</span>**

```python
# '일시' 열을 datetime 형식으로 변환
weather_21['일시'] = pd.to_datetime(weather_21['일시'])
weather_22['일시'] = pd.to_datetime(weather_22['일시'])

for df in [weather_21, weather_22]:
    df['year'] = df['일시'].dt.year
    df['month'] = df['일시'].dt.month
    df['day'] = df['일시'].dt.day
    df['hour'] = df['일시'].dt.hour
    df['weekday'] = df['일시'].dt.weekday
    
# datetime 컬럼 제거
weather_21.drop(columns='일시', inplace=True)
weather_22.drop(columns='일시', inplace=True)
```

```python
# 라벨 인코더 초기화
label_encoder = LabelEncoder()

# 라벨 인코딩을 적용할 열 선택
categorical_columns = ['지점명','운형(운형약어)']

# 라벨 인코딩을 적용할 데이터프레임 및 열 선택
for column in categorical_columns:
    weather_21[column] = label_encoder.fit_transform(weather_21[column])
    weather_22[column] = label_encoder.fit_transform(weather_22[column])
    
# 결측치가 10개 이상인 열을 삭제
threshold = 10  # 결측치의 최대 허용 개수
weather_21 = weather_21.dropna(axis=1, thresh=len(weather_21) - threshold)
weather_22 = weather_22.dropna(axis=1, thresh=len(weather_22) - threshold)
# 새로운 데이터프레임 확인
weather_21
```
